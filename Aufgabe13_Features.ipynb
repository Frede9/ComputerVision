```python
# import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

## Aufgabe A

### Performance

```python
# read image as grayscale
img = cv2.imread('boat.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print("Shape: ", img.shape)
plt.imshow(gray, cmap='gray');
```

#### SIFT - Performance

```python
img_sift = img.copy()
start = time.time()

sift = cv2.SIFT_create()
kp = sift.detect(gray,None)
img_sift1=cv2.drawKeypoints(gray,kp,img_sift)

end = time.time()
print(end - start)

cv2.imwrite('sift_keypoints_performance.jpg',img_sift1)
```

#### ORB - Performance

```python
img_orb = img.copy()
# Initiate ORB detector

start = time.time()
orb = cv2.ORB_create()
# find the keypoints with ORB
kp = orb.detect(img_orb,None)
# compute the descriptors with ORB
kp, des = orb.compute(img_orb, kp)
# draw only keypoints location,not size and orientation
img_orb1 = cv2.drawKeypoints(img_orb, kp, None, color=(0,255,0), flags=0)

end = time.time()
print(end - start)

cv2.imwrite('orb_keypoints_preformance.jpg',img_orb1)
```

### Translation

```python
M = np.float32([[1,0,100],[0,1,50]])

rows,cols = gray.shape

img_translated = cv2.warpAffine(img,M,(cols,rows))
img_translated_gray = cv2.cvtColor(img_translated,cv2.COLOR_BGR2GRAY)

print("Shape: ", img_translated.shape)
plt.imshow(img_translated_gray, cmap='gray');
```

#### SIFT-Translated

```python
img_sift = img_translated.copy()

sift = cv2.SIFT_create()
kp = sift.detect(img_translated_gray,None)
img_sift1=cv2.drawKeypoints(img_translated_gray,kp,img_sift)

cv2.imwrite('sift_keypoints_translated.jpg',img_sift1);
```

SIFT ist translations-invariant.

#### ORB-Translated

```python
img_orb = img_translated.copy()

orb = cv2.ORB_create()
# find the keypoints with ORB
kp = orb.detect(img_orb,None)
# compute the descriptors with ORB
kp, des = orb.compute(img_orb, kp)
# draw only keypoints location,not size and orientation
img_orb1 = cv2.drawKeypoints(img_orb, kp, None, color=(0,255,0), flags=0)

cv2.imwrite('orb_keypoints_translated.jpg',img_orb1);
```

ORB ist translations-invariant.

#### Sowohl SIFT als auch ORB sind nach Untersuchung der KeyPoints robust gegenüber Rotation, Skalierung und Translation. ORB jedoch weißt hier die deutlich bessere Performance auf und ist fast doppelt so schnell gegenüber SIFT.

## Aufgabe B

```python
img1 = cv2.imread('f1.jpg',0) 
img2 = cv2.imread('f2.jpg',0) 

# Initiate SIFT detector
sift = cv2.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)

# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)

# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])

img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
cv2.imwrite('matching.png',img3);
```
